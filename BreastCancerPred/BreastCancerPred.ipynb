```python
breastC['diagnosis'] = breastC.diagnosis.map({'M':1,'B':0})
```

```python
from sklearn.model_selection import train_test_split
X=breastC.iloc[:,2:]
y=breastC.iloc[:,1:2]
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.2, random_state=25)
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense, Activation
import numpy as np
from sklearn import metrics
from sklearn.model_selection import KFold
from scipy.stats import zscore


def create_network():
  model=Sequential()
  n_cols = X.shape[1]
  model.add(Dense(13, activation ='relu',input_shape=(n_cols,)))
  model.add(Dense(13, activation ='relu'))
  model.add(Dense(13, activation ='relu'))
  model.add(Dense(1, activation ='sigmoid'))
  model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
  return model
neural_network = KerasClassifier(build_fn=create_network, 
                                  epochs=10, 
                                  batch_size=100, 
                                  verbose=1)
cross_val_score(neural_network, X, y, cv=10).mean()
#model.fit(predictors, target, epoch=10)
#scores = cross_val_score(model, X, y, cv=10, scoring='accuracy')
  
    

```

```python
import matplotlib.pyplot as plt
breastC["diagnosis"].value_counts().plot.bar()
plt.show()
```

```python
history = neural_network.fit(X_train, y_train, epochs = 2000, validation_split = 0.15, verbose = 0)
history_dict=history.history
```

```python
# Predicting the Test set results
y_pred = neural_network.predict(X_test)
y_pred = (y_pred > 0.5)
```
