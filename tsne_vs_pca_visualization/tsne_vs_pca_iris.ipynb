#t-SNE

```python
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd
iris = load_iris()
X_tsne = TSNE(learning_rate=100).fit_transform(iris.data)
X_pca = PCA().fit_transform(iris.data)
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.shape
```

```python
X_tsne.shape
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')
ax1.scatter(X_tsne[:, 0], X_tsne[:, 1], c=iris.target)
ax1.set_title('tsne')
ax2.scatter(X_pca[:, 0], X_pca[:, 1], c=iris.target)
ax2.set_title('pca')
```

t-SNE puede ayudarnos a decidir si las clases son separables en alguna representación lineal o no lineal. Aquí podemos ver que las 3 clases del conjunto de datos de Iris se pueden separar con bastante facilidad. Incluso se pueden separar linealmente, lo que podemos concluir a partir de la baja dimensión del PCA.
en dimensiones altas y dominios no lineales PCA no es aplicable ni muchos otros algoritmos de aprendizaje no hacen buenas visualizaciones porque tratan de preservar la estructura global de los datos.
