#Clustering con el metodo de K-means.

variantes K-means 


1.   K-modas:
      utiliza moda en lugar de medias
      mejora variables categoricas
2.   K-medoids:
      es mas robusto al ruido y a los valores atipicos
      en lugar de usar centroides, utiliza individuos que se encuentran más centrados con respecto al cluster(medianas) (PAM, CLARA)
3.   Mini-batch K-Means:
      divide las bases de datos en subconjuntos que se van procesando en cada paso del algoritmo


tecnicas para determinar la cantidad de clusters 


1.   Codo de jambú: crear diferentes cantidades de clusters con k-means calculando que tan simmilares son los individuos dentro de los mismos y graficar
2.   Coeficiente de silueta
3.   Indice de dunn
4.   Davies-Bouldin Index
5.   Aplicar clustering jerarquico a una muestra tomada aleatoriamente

cuando no existe cohesión ni separación grandes 1 y 2.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
path = "/content/drive/MyDrive/caracteristicas de vinos.csv"
df = pd.read_csv(path, sep=",", encoding="ISO-8859-1", header=0, engine='python')
```

```python
df.info()
```

```python
df.head()
```

```python
vinos_variables=df.drop(['Vino'], axis=1)
```

```python
vinos_variables.describe()
```

```python
df.head()
```

#Normalizar valores

```python
vinos_norm=(vinos_variables-vinos_variables.min()) /(vinos_variables.max() -vinos_variables.min())
vinos_norm
```

```python
vinos_norm.describe()
```

#Busqueda de la cantidad óptima de clusters
Calculando que tan similares son los individuos dentro de los clusters

```python
ks = range(1, 6)
inertias = []

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(vinos_norm)
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

```python
clustering = KMeans(n_clusters=3)    
# Fit model to samples
clustering.fit(vinos_norm)
```

```python
df['Kmeans_Clusters'] = clustering.labels_
df.head()
```

#Visualización con PCA

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_vinos = pca.fit_transform(vinos_norm)
pca_vinos_df= pd.DataFrame(data = pca_vinos, columns = ['Componente_1', 'Componente_2'])
pca_nombres_vinos = pd.concat([pca_vinos_df, df[['Kmeans_Clusters']]], axis = 1)
pca_nombres_vinos
```

```python
fig = plt.figure(figsize = (6, 6))

ax = fig.add_subplot(1,1,1)
ax.set_xlabel('Componente 1', fontsize = 15)
ax.set_ylabel('Componente 2', fontsize = 15)
ax.set_title('Componentes Principales', fontsize = 20)

color_theme = np.array(["blue", "green", "orange"])
ax.scatter(x = pca_nombres_vinos.Componente_1, y = pca_nombres_vinos.Componente_2,
           c=color_theme[pca_nombres_vinos.Kmeans_Clusters], s = 50)
plt.show()
```
